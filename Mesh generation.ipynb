### Summoning the Biosphere *modulus* Mexico

As seen before, the biosphere object is the abstraction of the set of all occurrences of living beings in the GBIF database.
For instantiating the biosphere use:

```python
from gbif.models import Occurrence
biosphere = Occurrence.objects.all()
```

The content of this object is the entire GBIF databse. This implementation abstracts the GBIF occurrences in a way that the data is processed only if it's necessary. This techniques is called Lazy Evaluation.  This means that an expression, in this case a dataquery, is not evaluated when its gets bound to a variable, but when the evaluator is forced to produce the expression's value. This technique makes possible the abstraction of big data sets and ease the handling and representation of data structures. For further information read: 

`Hudak, P. (1989). "Conception, Evolution, and Application of Functional Programming-Languages." Computing Surveys 21(3): 359-411.`

*Caution:* any attempt to measure or retrieve data may cause a huge load of information.

#### Filtering the biosphere within the Mexican Border

```python
mexican_biosphere = biosphere.filter(geom__intersects=mex.geom)
```
